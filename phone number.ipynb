```python
def ValidphoneNumber(s):
  start =s.find("+") + len("+")
  end = s.find("-")
  code= s[start:end]
 
  start1=s.find("-") + len("-")
  end1=s.find(" ")
  Phnumber=s[start1:end1]
 
  countrycode=['1','2','3','91','98','255']  #you can add the country codes here
  for i in countrycode:
    if i == code and len(Phnumber) == 9 and Phnumber.isnumeric():   #string class method
      print(True)
      break
  else:
    print(False)
```

```python
ValidphoneNumber("+91-9061878989")
```
