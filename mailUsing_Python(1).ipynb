```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase                               #PACKAGES
```

```python
def sentmail(email_user , email_send , subject , body , fileName ):
  msg=MIMEMultipart()
  msg['From']=email_user
  msg['To']=email_send                  #assaign messages
  msg['Subject']=subject

  msg.attach(MIMEText(body,'plain'))      #attached 
  text=msg.as_string()                            #into string
  server=smtplib.SMTP('smtp.gmail.com',587)

  msg.attach(MIMEText(body,'plain'))      #attached 
  #filename='Suhair_Python-AI.pdf'   #filename
  filename = fileName
  attachment=open(filename,'rb')
  part=MIMEBase('application','octet-stream')
  part.set_payload((attachment).read())
  encoders.encode_base64(part)
  part.add_header('Content-Disposition',"attachment;filename="+filename)
  msg.attach(part)
  text=msg.as_string()                            #into string
  server=smtplib.SMTP('smtp.gmail.com',587)

  #if filename == "":
   # print("No files are attached")

  server.starttls()
  server.login(email_user,"duplicatemail@73")    #login email with password
  print("loggined successfuly")

  server.sendmail(email_user,email_send,text)
  server.quit()     
  print("mail sent successfully")
```

**SentOTP through mail**

```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase 
import random 
```

```python
email_user='suhairk58@gmail.com'
email_send='suhairtanur@gmail.com'              #data 
subject="Verification OTP"
```

```python
otp=random.randint(1000 , 9999)
otp = str(otp)
print(otp)
```

```python
msg=MIMEMultipart()
msg['From']=email_user
msg['To']=email_send                  #assaign messages
msg['Subject']=subject
body="Hi Your OTP number is" + otp
```

```python
msg.attach(MIMEText(body,'plain'))      #attached 
text=msg.as_string()                            #into string
server=smtplib.SMTP('smtp.gmail.com',587)

server.starttls()
server.login(email_user,"duplicatemail@73")    #login email with password
print("loggined successfuly")
```

```python
server.sendmail(email_user,email_send,text)
server.quit()     
print("mail sent successfully")
```
